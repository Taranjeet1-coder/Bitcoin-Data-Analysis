# Bitcoin History

Bitcoin, the first and most widely recognized cryptocurrency, has a fascinating history marked by significant milestones, price fluctuations, and technological developments. Here's an overview of Bitcoin's history:

## 1. Creation of Bitcoin (2008-2009)

- **2008: Whitepaper Release**  
  In **October 2008**, an anonymous person or group of people under the pseudonym **Satoshi Nakamoto** published the Bitcoin whitepaper titled *Bitcoin: A Peer-to-Peer Electronic Cash System*. This paper outlined the vision for a decentralized digital currency that would allow for peer-to-peer transactions without relying on banks or intermediaries.

- **2009: Bitcoin Network Goes Live**  
  On **January 3, 2009**, Nakamoto mined the first block of Bitcoin, known as the **Genesis Block** or **Block 0**, which contained the embedded message:  
  *"The Times 03/Jan/2009 Chancellor on brink of second bailout for banks."*  
  This message alluded to the financial crisis of 2008 and Bitcoin’s intention to challenge traditional financial systems.

  Bitcoin's initial price was **essentially $0**, and the first-ever transaction was between Nakamoto and computer scientist **Hal Finney** in January 2009, involving 10 BTC.

## 2. Early Years (2010-2012)

- **2010: The First Real-World Transaction**  
  On **May 22, 2010**, Bitcoin’s first known commercial transaction occurred when **Laszlo Hanyecz**, a programmer, paid **10,000 BTC** for **two pizzas**. This transaction is now famously known as "Bitcoin Pizza Day," and at the time, it was worth approximately **$25**.

- **2011: Bitcoin Hits $1**  
  In **February 2011**, Bitcoin reached the **$1** mark for the first time. It was gaining traction as a potential alternative to traditional currencies, especially among early tech enthusiasts and cryptography communities.

  By the end of 2011, Bitcoin’s value had risen to around **$31**, before falling back down due to various factors, including market instability and security breaches (such as hacks on Bitcoin exchanges).

## 3. Growth and Volatility (2013-2016)

- **2013: First Major Price Spike**  
  In **2013**, Bitcoin saw its first significant price surge, reaching **$266** in **April 2013**, before crashing to around **$50**. Later in the year, it began to rise again, reaching an all-time high of around **$1,200** by November 2013.
  
  In 2013, Bitcoin also began to gain some institutional attention, and countries like **Cyprus** were looking into Bitcoin during banking crises.

- **2014: Mt. Gox Hack**  
  The year **2014** was marked by the collapse of **Mt. Gox**, once the world’s largest Bitcoin exchange. In February 2014, Mt. Gox filed for bankruptcy after a hack resulted in the loss of **850,000 BTC** (worth around $450 million at the time). This event severely damaged Bitcoin’s reputation, leading to more regulatory scrutiny.

- **2015-2016: Recovery and Increasing Adoption**  
  In **2015**, Bitcoin experienced a steady recovery. The introduction of the **Bitcoin block size debate** and the debate around scalability became central in the Bitcoin community. Bitcoin’s infrastructure matured, and projects like **Litecoin** and **Ethereum** started to emerge alongside it.

  By **2016**, Bitcoin had regained much of its previous value and reached around **$700**.

## 4. The 2017 Boom (2017)

- **2017: Price Surge and Media Attention**  
  2017 was a pivotal year for Bitcoin as its price skyrocketed to **over $19,000** in December, driven by media hype, institutional interest, and the growing number of ICOs (Initial Coin Offerings) built on top of blockchain technology.

  This year also saw the rise of Bitcoin forks like **Bitcoin Cash**, as a result of disputes over the scalability of the Bitcoin network.

- **Bitcoin Futures**  
  Bitcoin futures trading began in **December 2017**, allowing institutional investors to speculate on Bitcoin's price. This led to further price volatility.

## 5. Crash and Recovery (2018-2020)

- **2018: The Bear Market**  
  After hitting its all-time high in December 2017, Bitcoin's price plummeted in early 2018, reaching lows of around **$3,000** by December 2018. The crash was partly due to regulatory concerns, market corrections, and the bursting of the ICO bubble.

- **2019-2020: Gradual Recovery and Institutional Interest**  
  By mid-2019, Bitcoin had risen back to around **$13,000**. However, the COVID-19 pandemic in 2020 spurred new interest in Bitcoin, with institutional players like **MicroStrategy**, **Tesla**, and large hedge funds beginning to buy Bitcoin as a hedge against inflation and currency devaluation.

## 6. The 2020-2021 Bull Run

- **2020: COVID-19 and Institutional Adoption**  
  In **2020**, Bitcoin's price began to surge again, breaking past **$20,000** in December, driven by institutional adoption, inflation concerns, and a broader acceptance of Bitcoin as "digital gold." Companies like **Tesla** and financial giants like **PayPal** began to integrate Bitcoin into their business models.

- **2021: New All-Time Highs**  
  In **2021**, Bitcoin reached new all-time highs, surpassing **$60,000** in April. The market was significantly influenced by institutional adoption, regulatory developments, and the launch of Bitcoin-based financial products like **Bitcoin ETFs**.

## 7. Volatility and Regulatory Scrutiny (2022-2023)

- **2022: Market Correction**  
  Bitcoin's price began a sharp decline from its 2021 highs in 2022. Bitcoin faced several challenges, including tightening monetary policy by central banks, particularly in the U.S., which led to increased volatility. By the end of 2022, Bitcoin's price hovered around **$16,000-$17,000**.

- **2023: Recovery and Market Resilience**  
  Bitcoin showed signs of recovery in **2023**, aided by the rise of institutional interest, developments in blockchain technology, and regulatory clarity in some countries. The market's overall trend remained bullish, though it remained susceptible to sudden corrections.

## 8. The Future of Bitcoin
The future of Bitcoin remains uncertain but promising. It is considered both a store of value (like digital gold) and a decentralized, censorship-resistant currency. However, it faces challenges from technological issues, regulatory scrutiny, competition from other cryptocurrencies (such as **Ethereum**), and the scalability debate.

Bitcoin’s path forward likely includes wider adoption in payment systems, continued institutional interest, and advancements in blockchain technology. Its role in the global economy, alongside traditional finance, will likely evolve as new use cases emerge.

```python
btc.rename(columns={'Price':'Close'}, inplace = True )
btc.rename(columns={'Vol.':'Volume'}, inplace = True )
btc
```

```python
# Ensure all Volume values are strings
btc['Volume'] = btc['Volume'].astype(str)

# Function to convert volume with 'k' , 'b' and 'm' into their respective numeric values
def convert_volume(volume):
    if 'M' in volume:
        return float(volume.replace('M', '')) * 1_000_000  # Convert millions
    elif 'K' in volume:
        return float(volume.replace('K', '')) * 1_000  # Convert thousands
    elif 'B' in volume:
        return float(volume.replace('B', '')) * 1_000_000_000  # Convert thousands
    else:
        try:
            return float(volume)  # Convert numeric strings to float
        except ValueError:
            return 0  # Handle invalid values gracefully

# Apply the conversion to the Volume column
btc['Volume'] = btc['Volume'].apply(convert_volume)

# Check the result
print(btc['Volume'])
```

```python
# Convert 'Close' column to numeric
btc['Close'] = pd.to_numeric(btc['Close'], errors='coerce')

# Check for any NaN values in 'Close' after conversion
print(btc['Close'].isna().sum())

# Drop rows with NaN values in 'Close' (if any exist)
btc = btc.dropna(subset=['Close'])

# Now plot the data
plt.figure(figsize=(12, 6))
plt.plot(btc['Date'], btc['Close'], label='Close Price', color='blue')
plt.title('Trend of Bitcoin Close Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.legend()
plt.grid()
plt.show()
```

## Moving Average of Bitcoin Prices

The **Moving Average (MA)** is a commonly used indicator in financial analysis. It helps smooth out short-term fluctuations in the price to reveal the **long-term trends**. Moving averages are often used to track the general direction of the market and to filter out noise from daily price movements.

### **Key Points from the Moving Average Chart**:
- **Smoothing Effect**: The moving average smooths the price data, making it easier to observe general trends. It is particularly useful for identifying underlying market conditions, such as bullish or bearish trends.
- **Short-term vs. Long-term**: A **short-term moving average** (e.g., 7-day) will be more sensitive to recent price changes, while a **long-term moving average** (e.g., 30-day) will capture broader market trends.
- **Sudden Spikes**: When the actual price moves sharply above or below the moving average, it indicates **significant price changes** or potential trend reversals. These spikes often indicate periods of high volatility or major market events.
- **Crossovers**: If the short-term moving average crosses above the long-term moving average, it might signal a **bullish trend**, and if it crosses below, it might indicate a **bearish trend**.

The moving average is an excellent tool for detecting trends, but it is important to consider other factors like market news or external influences when interpreting it.

```python
# Simple Moving Average (SMA) Over Time for Monthly Data
def plot_moving_average(data, window_size=3, start_date=None, end_date=None):
    # Apply filters
    filtered_data = data.copy()
    
    if start_date:
        filtered_data = filtered_data[filtered_data['Date'] >= pd.to_datetime(start_date)]
    if end_date:
        filtered_data = filtered_data[filtered_data['Date'] <= pd.to_datetime(end_date)]
    
    # Check if there are enough data points for monthly data
    if len(filtered_data) < window_size:
        print("Not enough data points for the given window size.")
        return
    
    # Calculate the Simple Moving Average (SMA)
    filtered_data['SMA'] = filtered_data['Close'].rolling(window=window_size).mean()

    # Plot: BTC Closing Price and SMA Over Time
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=filtered_data, x='Date', y='Close', label='Closing Price', color='blue')
    sns.lineplot(data=filtered_data, x='Date', y='SMA', label=f'{window_size}-month SMA', color='orange')
    plt.title(f'BTC Closing Price and {window_size}-month SMA Over Time (Filtered)', fontsize=16)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Price (USD)', fontsize=12)
    plt.legend()
    plt.grid()
    plt.show()

# Example usage with monthly data
plot_moving_average(btc, window_size=3, start_date='2023-01-01', end_date='2024-01-01')
```
